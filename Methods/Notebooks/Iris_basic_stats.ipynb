# Iris example: data loading, basic stats and correlation

```python
import pandas as pd
import seaborn as sn
```

## Read and examine data

```python
df = pd.read_csv('iris.csv')
df.head(10)
```

## Check and convert types

```python
df.dtypes
```

```python
df['species'] = df['species'].astype(pd.api.types.CategoricalDtype(ordered=False))
df.dtypes
```

```python
df.describe()
```

## Correlations

```python
corrmatrix = df.corr()
print(corrmatrix)
```

```python
sn.heatmap(corrmatrix, vmin=-1.0, vmax=1.0, cmap="RdBu")
```

## Scatterplots per species

```python
sn.pairplot(df, hue="species")
```
