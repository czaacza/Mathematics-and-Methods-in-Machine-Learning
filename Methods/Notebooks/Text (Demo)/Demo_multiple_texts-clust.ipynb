```python
import nltk
import urllib.request
import math
import os
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

#nltk.download()


def sentiment(path):
    text = open(path,'r').read()
    tokens = text.split()
    tokens=[token.lower() for token in tokens if token.isalpha()]
    sw = stopwords.words('english')
    ctokens = tokens
    for token in tokens:
        if token in sw:
            ctokens.remove(token)            
    lemmatizer = WordNetLemmatizer()
    for i in range(len(ctokens)):
        ctokens[i] = lemmatizer.lemmatize(ctokens[i])
    pos = open('C:\\Users\\olliv\\Documents\\cogsys\\positive-words.txt','r').read().split()
    neg = open('C:\\Users\\olliv\\Documents\\cogsys\\negative-words.txt','r').read().split()
    countpositive = countnegative = countneutral = counttotal = 0
    for token in ctokens:
        counttotal = counttotal + 1
        cat = ""
        if token in pos:
            countpositive = countpositive + 1
        elif token in neg:
            countnegative = countnegative + 1
        else:
            countneutral = countneutral + 1

    res = {}
    res["positive"] = countpositive/counttotal
    res["negative"] = countnegative/counttotal
    res["neutral"] = countneutral/counttotal
    res["positiveness"] = math.log(countpositive/countnegative)
    res["affection"] = 1-countneutral/counttotal    
    
    return res


def freqs(path):
    text = open(path,'r').read()
    tokens = text.split()
    tokens=[token.lower() for token in tokens if token.isalpha()]
    sw = stopwords.words('english')
    ctokens = tokens
    for token in tokens:
        if token in sw:
            ctokens.remove(token)            
    lemmatizer = WordNetLemmatizer()
    for i in range(len(ctokens)):
        ctokens[i] = lemmatizer.lemmatize(ctokens[i])
        
    freq = nltk.FreqDist(tokens) 
    return freq

#freqs('C:\\Users\\olliv\\Documents\\cogsys\\speeches\\2009-Obama.txt')
    
```

```python
directory = 'C:\\Users\\olliv\\Documents\\cogsys\\speeches'

for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        res = sentiment(os.path.join(directory, filename))
        print("%40s %.3f %.3f" % (filename, res['positiveness'], res['affection']))
```

```python
directory = 'C:\\Users\\olliv\\Documents\\cogsys\\speeches'
freqdata = []
speechname = []
for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        speechname.append(filename)
        freqdata.append(freqs(os.path.join(directory, filename)))
        
print(freqdata)
```

```python
from sklearn.feature_extraction import DictVectorizer
import pandas as pd
import numpy as np

vec = DictVectorizer()

narray = vec.fit_transform(freqdata).toarray()
fnames = vec.get_feature_names()
df = pd.DataFrame(narray)
df.columns=fnames
df = df.iloc[:,1:]
df

```

```python
from sklearn.preprocessing import StandardScaler
from sklearn import cluster

scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df))
df_scaled.head(10)
```

```python
clu = cluster.AgglomerativeClustering(n_clusters=10)
clu.fit(df_scaled)

labels = clu.labels_

for i in range(len(labels)):
    print (labels[i], speechname[i])
```
