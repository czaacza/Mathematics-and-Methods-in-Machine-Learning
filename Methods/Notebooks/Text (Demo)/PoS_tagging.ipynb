```python
import nltk
#nltk.download('popular')
```

```python
text = """

"""
tokens = nltk.word_tokenize(text)
tokens
```

```python
tagged = nltk.pos_tag(tokens)
tagged
```

```python
grammar = "Name: {(<NNP|NNPS>+<IN>*<NNP|NNPS>+)|<NNP|NNPS>+}"
chunk_parser = nltk.RegexpParser(grammar)
tree = chunk_parser.parse(tagged)
tree.draw()
```
