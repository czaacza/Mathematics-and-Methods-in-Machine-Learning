# Generate data set

```python
import pandas as pd
import numpy as np



dict = {
    "id": [1,2,3,4,5],
    "age": [20,34,76,58,44],
    "gender": ['F', 'M', 'M', 'M', 'F']
}


df = pd.DataFrame(dict)
df
```

# Add new variable

```python
df["height"] = [158,164,170,165,161]
df
```

# Add rows

```python
df.append(pd.Series([6, 32, "F", 173], index=df.columns), ignore_index=True)
```

# Combine data frames based on id

```python
dict2 = {
    "id": [1,2,3,5],
    "weight": [60,64,80,82],
}


df2 = pd.DataFrame(dict2)
df2
```

```python
df = df.merge(df2, on="id", how="left")
df
```

# Filter rows

```python
df[df["age"] >= 40]
```

# Select colums

```python
df3 = df[["id","age", "height"]]
df3
```

# Drop variables

```python
df4 = df.drop(["age","gender"], axis=1)
df4
```

# Batch modify
Add 1 year to everybody's age

```python
df["age"] = df["age"] + 1
df
```

Add 2 kg of weight to all males

```python
df.loc[(df["gender"]=="M"),"weight"] = df.loc[(df["gender"]=="M"),"weight"] + 2 
df
```

# Drop rows
Drop row where id = 4.
First generate an index of the rows to be dropped.
