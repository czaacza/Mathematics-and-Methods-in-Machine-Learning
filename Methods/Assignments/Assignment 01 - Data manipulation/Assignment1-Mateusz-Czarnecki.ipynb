## 2.

```python
data_dict = {
    "Weight": [92, 70, 58, 99, 55, 76, 62, 92, 71, 70, 77, 79],
    "Exercise": [6, 6, 6, 2, 8, 4, 6, 6, 5, 6, 4,4 ],
    "Cholesterol": [4.8, 5.1, 6.4, 6.5, 2.3, 5.7, 4.2, 6.9, 4.8, 4.8, 7.7, 5.7],
    "Income": [2060, 2660, 2530, 1740, 3520, 3750, 2720, 3130, 2100, 3340, 2430, 2700],
    "Happiness": [49, 36, 49, 28, 77, 55, 43, 39, 54, 29, 53, 47],
    "Birthyear": [1953, 1955, 1939, 1942, 1989, 1937, 1979, 1905, 1995, 1966, 1938, 1993]
}
```

```python
data_frame = pd.DataFrame(data_dict)
```

```python
data_frame
```

### a) Basic statistics

```python
statistics = data_frame.describe()
statistics
```

### b) Average Income

```python
average_income = statistics.loc['mean']['Income']

for index, row in data_frame.iterrows():
    if row["Income"] < average_income:
        print(f'Person {index}: below average income.')
    elif row["Income"] > average_income:
        print(f'Person {index}: above average income.')
    else:
        print(f'Person {index}: equal to the average income.')
        
```
