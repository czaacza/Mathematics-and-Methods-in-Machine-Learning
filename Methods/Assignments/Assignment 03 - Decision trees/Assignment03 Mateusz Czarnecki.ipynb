```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn import preprocessing

import graphviz
import pydotplus
from sklearn.tree import export_graphviz

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from IPython.display import Image

# read data
df = pd.read_csv('data/phishing.csv', sep=';')

# drop rows with missing values
df.dropna(axis=0, how='any', inplace=True)
df.head(10)
```

```python
colnames = df.columns
colnames
```

```python
X = df.drop(['Result'], axis=1)
Y = df.loc[:, 'Result']
# cat_features = 

# encoder = preprocessing.OneHotEncoder().fit(features[cat_features])
```

```python
classifier = tree.DecisionTreeClassifier(max_depth=5)
model = classifier.fit(X,Y)
```

```python
# dot_data = export_graphviz(classifier, out_file=None, feature_names=X.columns, filled=True)

# graph = pydotplus.graph_from_dot_data(dot_data)

# Image(graph.create_png())
```

```python
# visualize
text_representation = tree.export_text(classifier)
print(text_representation)
```

```python
# predict
Y_pred = classifier.predict(X)
# output confusion matrix
cm = confusion_matrix(Y, Y_pred)
print("Confusion matrix:\n",cm)
accuracy = accuracy_score(Y, Y_pred)
print("\nAccuracy calculated from the training set = %.3f" % (accuracy))
print(classification_report(Y, Y_pred, target_names=['no', 'yes']))
```

```python
# cross-validate
# number of folds
k = 10
scores = cross_val_score(estimator=classifier,
X=X,
y=Y,
scoring="accuracy",
cv=k)
print("Accuracies from %d individual folds:" % k)
print(scores)
print("Accuracy calculated using %d-fold cross validation = %.3f" % (k, scores.mean()))
```
