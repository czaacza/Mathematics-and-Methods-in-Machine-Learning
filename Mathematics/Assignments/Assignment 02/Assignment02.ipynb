### 2

```python
x = np.arange(1,11)
y = np.array([14,12,34,24,43,38,41,57,55,50])

plt.plot(x,y, 'ro')
```

```python
X = np.concatenate((np.ones((10,1)), x.reshape(10,1)) , axis=1)
Y = y.reshape((10,1))

M = la.inv(X.T.dot(X)).dot(X.T).dot(Y)

yy = X.dot(M)

plt.plot(x, yy)
plt.plot(x,y ,'ro')
```

### 3

```python
X = np.concatenate((np.ones((10,1)), x.reshape(10,1), x.reshape(10,1)**2), axis=1)
Y = y.reshape((10,1))

M = la.inv(np.dot(X.T, X)).dot(X.T).dot(Y)

xx = np.linspace(1, 10, 1000)
XX = np.concatenate((np.ones((1000,1)), xx.reshape(1000,1), xx.reshape(1000,1)**2), axis=1)


yy = XX.dot(M)
plt.plot(xx, yy)
plt.plot(x, y, 'ro')
plt.show()
```

### 4

```python
coef = np.polyfit(x,y, 4)
p = np.poly1d(coef)

xx = np.linspace(1, 10, 1000)
yy = p(xx)

plt.plot(x, y, 'ro')
plt.plot(xx, yy)
plt.show()
```

The polynomials of order higher than 3 or 4 do not fit the data better. The function then starts overfitting the data. For the prediction, I'd use the third or fourth degree polynomial
